```python
## sub plot 

## it take three argument row,columns,Index

plt.subplot()
```

```python
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)

plt.plot(dataset['age'],dataset['bmi'])

plt.subplot(1,2,2)
plt.plot(dataset['age'],dataset['charges'])
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(dataset['bmi'],dataset['age'])

plt.subplot(2,2,2)
plt.plot(dataset['charges'],dataset['age'])

plt.subplot(2,2,3)
plt.plot(dataset['region'],dataset['age'])

plt.subplot(2,2,4)
plt.plot(dataset['bmi'],dataset['age'])
```

```python
year=[2010,2012,2013,2014,2015]
yiled_apples=[0.895,0.919,0.926,0.929,0.931]
plt.plot(year,yiled_apples)
```

```python
plt.plot(year,yiled_apples)
plt.xlabel('year')
plt.ylabel('yiled of apples ')
```

```python
year=[2010,2012,2013,2014,2015]
yiled_apples=[0.895,0.919,0.926,0.929,0.931]
yiled_oranges=[0.845,0.939,0.966,0.959,0.981]
plt.plot(year,yiled_apples,marker = 'o')
plt.plot(year,yiled_oranges,marker='*')
```
